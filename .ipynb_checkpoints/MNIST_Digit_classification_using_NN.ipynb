**MNIST Handwritten Digit Clasification using (Neural Network)**

Importing the Dependencies

```python
import numpy as np  # importing numpy for performing a large number of mathematical operations, working with matrices,arrays
import math
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from PIL import Image
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras

from tensorflow.keras.datasets import mnist   # type: ignore
from sklearn.metrics import confusion_matrix

```

Loading the MNIST data from keras.datasets

```python
(X_train, Y_train), (X_test, Y_test) =  mnist.load_data()
```

```python
type(X_train)
```

```python
# shape of the numpy arrays
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

Training data = 60,000 Images

Test data = 10,000 Images

Image dimension  --> 28 x 28

Grayscale Image  --> 1 channel

```python
# printing the 10th image

print(X_train[10])
```

```python
print(X_train[10].shape)
```

```python

plt.imshow(X_train[25])   # displaying the image
plt.show()

print(Y_train[25])  # print the corresponding label
```

Image Lables

```python
print(Y_train.shape, Y_test.shape)
```

```python
# unique values in Y_train
print(np.unique(Y_train))

# unique values in Y_test
print(np.unique(Y_test))
```

All the images have the same dimensions in this dataset, If not, we have to resize all the images to a common dimension

```python
# scaling the values

X_train = X_train/255
X_test = X_test/255
```

```python
# printing the 10th image

print(X_train[10])
```

Building the Neural Network

```python
# setting up the layers of the Neural  Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(50, activation='relu'),
                          keras.layers.Dense(50, activation='relu'),
                          keras.layers.Dense(10, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network

model.fit(X_train, Y_train, epochs=10)
```

Training data accuracy = 98.9%

**Accuracy on Test data:**

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(accuracy)
```

Test data accuracy = 97.1%

```python
print(X_test.shape)
```

```python
# first data point in X_test
plt.imshow(X_test[0])
plt.show()
```

```python
print(Y_test[0])
```

```python
Y_pred = model.predict(X_test)
```

```python
print(Y_pred.shape)
```

```python
print(Y_pred[0])
```

model.predict() gives the prediction probability of each class for that data point

```python
# converting the prediction probabilities to class label

label_for_first_test_image = np.argmax(Y_pred[0])
print(label_for_first_test_image)
```

```python
# converting the prediction probabilities to class label for all test data points
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

Y_test  -->  True labels

Y_pred_labels  -->  Predicted Labels

Confusin Matrix

```python
conf_mat = confusion_matrix(Y_test, Y_pred_labels)  
```

```python
print(conf_mat)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
plt.ylabel('True Labels')
plt.xlabel('Predicted Labels')
```
