```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.preprocessing import StandardScaler
```

```python
data =pd.read_csv("train.csv")
```

```python
data = np.array(data)
m, n = data.shape   #m = Number of rows (total data samples)   n -> Number of columns (features + label in MNIST).
np.random.shuffle(data) # shuffle before splitting into dev and training sets

data_dev = data[0:1000].T
Y_dev = data_dev[0]   #first row (after transposing) contains the labels (digits 0-9).
X_dev = data_dev[1:n] # remaining rows (from index 1 to n-1) contain pixel values (features).
X_dev = X_dev / 255.  #scaling them between 0 and 1.

data_train = data[1000:m].T    # training data
Y_train = data_train[0]
X_train = data_train[1:n]
X_train = X_train / 255.
_,m_train = X_train.shape
```

### Our NN will have a simple two-layer architecture. Input layer  a[0]will have 784 units corresponding to the 784 pixels in each 28x28 input image. A hidden layer  a[1]
### will have 10 units with ReLU activation, and finally our output layer  a[2]
###  will have 10 units corresponding to the ten digit classes with softmax activation.

```python
def init_params():
    W1 = np.random.rand(10, 784) - 0.5
    b1 = np.random.rand(10, 1) - 0.5
    W2 = np.random.rand(10, 10) - 0.5
    b2 = np.random.rand(10, 1) - 0.5
    return W1, b1, W2, b2

def ReLU(Z):  #Z-> The input matrix (weighted sum of inputs + bias).
    return np.maximum(Z, 0)

def softmax(Z):  # Softmax converts raw scores into probabilities and Used in the output layer of classification models.
    A = np.exp(Z) / sum(np.exp(Z)) 
    return A
 
 
# Forward propagation is the process of passing input data through the network layer by layer to obtain the output predictions.   
def forward_prop(W1, b1, W2, b2, X):
    Z1 = W1.dot(X) + b1   # Z1 will give Raw scores before activation.
    A1 = ReLU(Z1)
    Z2 = W2.dot(A1) + b2
    A2 = softmax(Z2)   #Converts Z2 values into probabilities.
    return Z1, A1, Z2, A2

def ReLU_deriv(Z):  # this will used in backpropagation to update weights.
    return Z > 0

def one_hot(Y):  #Converts labels into one-hot encoded format for multi-class classification.
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y

# Backward propagation computes gradients to adjust weights and biases, minimizing the loss function.
def backward_prop(Z1, A1, Z2, A2, W1, W2, X, Y):
    one_hot_Y = one_hot(Y)
    dZ2 = A2 - one_hot_Y  #Error = Predicted - Actual.
    dW2 = 1 / m * dZ2.dot(A1.T)
    db2 = 1 / m * np.sum(dZ2)
    dZ1 = W2.T.dot(dZ2) * ReLU_deriv(Z1)
    dW1 = 1 / m * dZ1.dot(X.T)
    db1 = 1 / m * np.sum(dZ1)
    return dW1, db1, dW2, db2  


def update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
    W1 = W1 - alpha * dW1  #dW1 tells us how much W1 should change.
    b1 = b1 - alpha * db1    
    W2 = W2 - alpha * dW2  
    b2 = b2 - alpha * db2    
    return W1, b1, W2, b2

```

```python
def get_predictions(A2):  #extracting the predicted class with the highest probability.
    return np.argmax(A2, 0)

def get_accuracy(predictions, Y):    #calculateing the percentage of correct predictions.
    print(predictions, Y)
    return np.sum(predictions == Y) / Y.size

def gradient_descent(X, Y, alpha, iterations):
    W1, b1, W2, b2 = init_params()
    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)
        dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W1, W2, X, Y)
        W1, b1, W2, b2 = update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)
        if i % 10 == 0:  #Monitors accuracy every 10 iterations to track progress.
            print("Iteration: ", i)  
            predictions = get_predictions(A2)
            print(get_accuracy(predictions, Y))
    return W1, b1, W2, b2
```

```python
W1, b1, W2, b2 = gradient_descent(X_train, Y_train, 0.10, 500)
```

```python
def make_predictions(X, W1, b1, W2, b2):  #Passes an image through the network to get the predicted digit.
    _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
    predictions = get_predictions(A2)
    return predictions

def test_prediction(index, W1, b1, W2, b2):
    current_image = X_train[:, index, None]
    prediction = make_predictions(X_train[:, index, None], W1, b1, W2, b2)
    label = Y_train[index]
    print("Prediction: ", prediction)
    print("Label: ", label)
    
    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()
```

```python
test_prediction(0, W1, b1, W2, b2)
test_prediction(1, W1, b1, W2, b2)
test_prediction(2, W1, b1, W2, b2)
test_prediction(3, W1, b1, W2, b2)
```

```python
dev_predictions = make_predictions(X_dev, W1, b1, W2, b2)
get_accuracy(dev_predictions, Y_dev)
```
